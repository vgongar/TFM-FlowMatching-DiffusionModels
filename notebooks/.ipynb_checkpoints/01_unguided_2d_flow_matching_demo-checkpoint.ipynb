# Easy Unguided Flow Matching

In this notebook we will learn a smiley-type distribution using Flow Matching

```python
class VectorFieldNet(nn.Module): # Neural network for the vector field
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(3, 128), # Input: 2 coordinates (x,y) + 1 time (t) = 3
            nn.SiLU(), 
            nn.Linear(128, 128),
            nn.SiLU(),
            nn.Linear(128, 128),
            nn.SiLU(),
            nn.Linear(128, 2) # Output: Velocity (vx, vy)
        )

    def forward(self, x: torch.tensor, t):
        # Concatenate inputs x and t into a single column vector
        t_vec = t.view(-1, 1)
        x_input = torch.cat([x, t_vec], dim=-1)
        return self.net(x_input)
```

```python
# Load the model and training configuration
DEVICE = "cpu" #torch.device("cuda" if torch.cuda.is_available() else "cpu")
BATCH_SIZE = 500 
EPOCHS = 1000
```

```python
# To access batches efficiently
dataloader = torch.utils.data.DataLoader(data, batch_size=BATCH_SIZE, shuffle=True)
```

```python
mse_loss_fn = torch.nn.MSELoss(reduction='mean')
```

```python
def train_net(model, dataloader, optimizer, epochs=10):
    metrics_dic = {'rec_loss': []}
    for epoch in range(1,epochs+1):
        rec_loss_sum = 0 
        pbar = tqdm(enumerate(dataloader), total=len(dataloader))
        pbar.set_description(f"Epoch {epoch} - Training")
        for batch_idx, batch in pbar:
            # Sample a batch
            z = batch.to(DEVICE).float()

            # Sample random uniform time t ~ [0, 1]
            t = torch.rand(BATCH_SIZE, device=DEVICE).view(-1, 1) # transpose

            # Sample gaussian noise
            eps = torch.randn_like(z).to(DEVICE)

            # Equivalent to sampling from N(tz,(1-t)^2*Id)
            x = t * z + (1-t) * eps

            # Propagate the neural network
            output = model(x, t)

            # Conditional target
            target = z - eps

            # Evaluate the loss function
            rec_loss = mse_loss_fn(output, target)

            # Calculate gradient with backpropagation
            optimizer.zero_grad()
            rec_loss.backward()
            optimizer.step()

            # Training information
            rec_loss_sum += rec_loss.item()
            pbar.set_postfix({'rec_loss' : rec_loss_sum / (batch_idx+1)})
        rec_loss_avg = rec_loss_sum / len(dataloader)
        metrics_dic['rec_loss'].append(rec_loss_avg)
    return model, metrics_dic
```

```python
model = VectorFieldNet().to(DEVICE)
opt = torch.optim.Adam(model.parameters())
model.train()
model, _ = train_net(model, dataloader, opt, epochs=EPOCHS)
```

```python
@torch.no_grad() # Don't calculate gradients in this function
def generate_euler(model, n=2000, steps=100):
    """
    Simulates n points using Euler's integration method.
    """
    model.eval() # We set the network in evaluation mode instead of training

    noise = torch.randn(n, 2).to(DEVICE)
    x_t = [noise]
    h = 1.0 / steps

    # Euler Method
    # We can feed the network all sampled points at once
    # So Euler Method will be calculated on every row (each row is a sample point).
    for i in range(steps):
        
        t_val = i / steps
        t = torch.full((n,), t_val, device=DEVICE) # (n,1)
        
        x = x_t[-1] # (n,2)

        # Evaluate the net to get the velocities
        v = model(x, t) 
        
        # Update positions via Euler integration and add it to the trajectory
        x_t.append(x + v * h) # Euler
        
    return np.array(x_t)
```

```python
steps = 100
x_t = generate_euler(model, steps = steps)
```

```python
n_plots = 10
rows = 2
cols = 5

plt.figure(figsize=(15, 6))

indices = np.linspace(0, len(x_t)-1, n_plots, dtype=int)

for i, idx in enumerate(indices):
    plt.subplot(rows, cols, i + 1)

    # For title and color purposes
    t_curr = idx / (len(x_t)-1)
    color = plt.cm.viridis(t_curr) # colormap
    
    plt.scatter(x_t[idx][:, 0], x_t[idx][:, 1], s=1, alpha=0.5, color=color)
    
    plt.title(f"t={t_curr:.2f}")
    plt.xlim(-2.5, 2.5)
    plt.ylim(-2.5, 2.5)

    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
n_steps, n_samples, _ = x_t.shape
steps = np.repeat(np.arange(n_steps), n_samples)
ids = np.tile(np.arange(n_samples), n_steps)
coords = x_t.reshape(-1, 2)

export = pd.DataFrame({
    'step': steps,
    'ids': ids,
    'x': coords[:, 0],
    'y': coords[:, 1]
})

export.to_csv('trayectoria.csv', index = False)
```
